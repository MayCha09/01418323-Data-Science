## The Data

ใน exercise นี้ เราจะใช้ titanic data set ซึ่งเป็น Data ที่มีชื่อเสียงเป็นที่นิยม ซึ่งมีให้อยู่ใน library ของ seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

# Exercises

** ให้นำ data ของ titanic ที่เป็น dataframe ไปใช้ทำการ plot ให้ตรงกับภาพดังต่อไปนี้**

** *Note! เพื่อไม่ให้ภาพที่ถูกต้องหาย อย่าลืม insert cell ขึ้นมาข้างบนใหม่มาขั้นก่อนทำการ run code จะได้ไม่ทำให้เกิดกันทับซ้อน (overwrite)* **


**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
sns.jointplot(data = titanic , x = 'fare', y = 'age')
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
sns.histplot(titanic['fare'],bins = 26, color = '#F28585')
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
sns.boxplot(data = titanic, x = 'class' , y = 'age' ,palette = "husl")
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
sns.swarmplot(data = titanic, x = 'class' , y = 'age' , palette = 'Set2')
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
sns.countplot(data = titanic , x = 'sex')
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
sns.heatmap(titanic.corr() , cmap = 'coolwarm')
```

```python
sns.heatmap(titanic[['survived','pclass','age','sibsp','parch','fare','adult_male','alone']].corr(), cmap = 'coolwarm')
```

**จงทำการ plot ให้เหมือนกับรูปตัวอย่างด้านล่างนี้**

```python
g = sns.FacetGrid(data = titanic, col="sex"  )
g.map ( sns.histplot , 'age', bins = 10, alpha = 1)
```
